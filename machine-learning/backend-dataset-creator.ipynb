```python
import pandas as pd
```

```python
df = pd.read_csv('combined-dataset/combined_datasetV2.csv')
df.dropna(subset=['rating'], inplace=True)
```

```python
filtered = df[df['rating-count'] >= 50]
filtered.reset_index(drop=True, inplace=True)
filtered = filtered[filtered['rating'] >= 3.5]
```

```python
filtered.dropna(subset=['primary-type'], inplace=True)
filtered.dropna(subset=['review 1', 'rev'])
filtered
```
